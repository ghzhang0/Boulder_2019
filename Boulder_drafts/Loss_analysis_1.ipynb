```python
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
```

```python
import seaborn as sns
sns.set_palette("coolwarm", 10)
```

```python
Ns = [50, 100, 200, 300]
Rexps = [0.0, 0.1, 0.25, 0.3, 0.4, 0.5, 0.6, 0.7, 0.75, 0.8, 0.85, 0.9, 1.0]

N = 300
plt.figure(figsize=(14, 8))
for Rexp in Rexps[-1:]:
    df = pd.read_csv('Data_22july/loss_{}_{}.dat'.format(N, Rexp), sep=",", header=None)
    loss = np.array(df).T[0]
    plt.plot(loss)
plt.show()
```

```python
loss_all = np.zeros((len(Ns), len(Rexps)))

for i, N in enumerate(Ns):
    for j, Rexp in enumerate(Rexps):
        print(N, Rexp)
        df = pd.read_csv('Data_22july/loss_{}_{}.dat'.format(N, Rexp), sep=",", header=None)
        loss = np.array(df).T[0][-1]
        loss_all[i][j] = loss
        
```
